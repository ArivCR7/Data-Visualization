```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
import urllib.request
```

# Tabulation

```
url = "https://api.covid19india.org/states_daily.json"
urllib.request.urlretrieve(url, 'data.json');
```

```
df = pd.read_json('data.json')
df.head()
```

```
with open('data.json', 'r') as f:
  json_data = json.load(f)
```

```
type(json_data)
```

```
json_data.keys()
```

```
json_data = json_data['states_daily']
```

```
type(json_data)
```

```
json_data[:5]
```

```
df = pd.json_normalize(json_data)
df.head()
```

```
df['date'] = pd.to_datetime(df['date'])
df.shape
```

```
df = df[df.status == 'Confirmed']
df.shape
```

```
df.tail(7)
```

```
df.drop('status', axis=1, inplace=True)
```

```
df.set_index('date', inplace=True)
df.tail(7)
```

```
df.info()
```

```
df = df.apply(pd.to_numeric)
df.info()
```

```
df.tail()
```

# Table Styling

```
df = df.tail(10)
df
```

Let's highlight the numbers which are negative

```
def mark_negative_red(x):
  color = 'red' if x < 0 else 'black'
  return "color: "+color
```

```
df.style.applymap(mark_negative_red)
```

Thank god, no negatives!

Highlight the max count in each column as red and min count in green

```
df.style.highlight_max(color='red').highlight_min(color='green')
```

Looks like few columns are not needed. Removing those..

```
df.drop(['dd','ld','un','tt'], axis=1, inplace=True)
```

```
df.style.highlight_max(color='red').highlight_min(color='green')
```

The above gives an intuition on when min and max falls for each state in the last 10 days.

In the last 10 days, let's indicate max count in each column along with highlighting the max count for each day. To bold the max count in each column, let's write a function. 

```
def bold_max_count(x):
  ismax = (x == x.max())
  return ['font-weight:bold' if y else '' for y in ismax]
```

```
df.style.apply(bold_max_count).highlight_max(color='red', axis=1)
```

Maharashtra signals alarm by showing max count in the last 10 days. Tamilnadu reaches max count of 5688 on Oct 1.

Now let's apply color gradient which gives much more appealing.

```
df.style.background_gradient(cmap='Reds', axis=1)
```

Only few states seems to be more alarming. Let's subset those:

```
interested_states = ['ap','ka','kl','mh','tn','up']
```

```
df.style.apply(bold_max_count).background_gradient(cmap='Reds', axis=1, subset=interested_states)
```

Let's plot a growing bar for each state. 

```
df.style.bar()
```

```
df.style.bar(subset=interested_states)
```

```
df[interested_states].style.bar()
```
